```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/OCO-2 dataset.csv')
df.head()
```

```python
df=df.drop(['reg05','reg07','reg08','reg09','reg10','reg11','reg12','reg13'],axis=1)
```

```python
df
```

```python
df=df.drop(['reg04','reg14','reg15'],axis=1)
```

```python
df
```

```python
df.tail()
```

```python
mask = df['site'] == 'text'

# select all rows except the ones that contain 'Coca Cola'
df = df[~mask]

# print the resulting DataFrame
print(df)
```

```python
df.head()
```

```python
df=df.drop('reg03',axis=1)
```

```python
df
```

```python
for column in df.columns[4:]:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Convert the date column to a proper date format
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Remove rows with any null values
df = df.dropna()
```

```python
df
```

```python
plt.figure(figsize=(14, 7))
```

```python
for column in df.columns[4:]:
    plt.plot(df['date'], df[column], label=column)
plt.xlabel('Date')
plt.ylabel('Values (ppm)')
plt.title('Time Series of reg01 to reg17')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], c='blue', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Longitude vs Latitude')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.iloc[:, 4:].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Correlation Matrix for reg01 to reg17')
plt.show()
```

```python
sns.pairplot(df.iloc[:, 4:])
plt.suptitle('Pair Plot of reg01 to reg17', y=1.02)
plt.show()
```

```python
df.iloc[:, 4:].hist(bins=15, figsize=(15, 10), layout=(4, 5))
plt.suptitle('Histograms of reg01 to reg17', y=1.02)
plt.show()
```
